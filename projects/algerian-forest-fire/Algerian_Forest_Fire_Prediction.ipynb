```python
from logging import info

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.pylabtools import figsize
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"region"] = 0
df.loc[122:,"region"] = 1
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.info()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtypes == 'object']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv('Algerian_Forest_Fire_Cleaned_dataset.csv',index=False)
```

## EDA

```python
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Encode the classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains( 'not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
# plot density
df_copy.hist(bins=50,figsize=(50,50))
plt.show()
```

```python
## percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
classlabels = ["fire","not fire"]
plt.figure(figsize=(12,6))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Percentage of fires in each class')
plt.show()
```
